# X-ray luminosity functions of Clusters in the ERASS survey

Import the necessary packages and functions

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import sys
sys.path.append('../../')
from scripts.luminosity_functions import ClustersXrayLuminosityFunction
from scripts.utils import set_mpl
from scipy import integrate
import pyccl as ccl
set_mpl()

path2plots = './plots/'
```

## Selection function of clusters

```python
cluster_xlf_tests = ClustersXrayLuminosityFunction()

fig,  ax =  plt.subplots( figsize = (12,12))
slim = 4.4e-14 #limiting flux in cgs
zarr = np.linspace(0.01, 1.3, 50)
mmins = cluster_xlf_tests.Mmin(slim, zarr)
ax.semilogy(zarr, mmins*0.7, '-',label=f'Mmin, slim = {slim}', lw = 3)
ax.set_xlabel('z')
ax.set_ylabel('$M_{min}$, Msun/h')
ax.legend()
```

## Plot distribution of Clusters  in redshift

### dNdz

```python
Slim_ext_soft = 4.4e-14
clusters_xlf = ClustersXrayLuminosityFunction()
h = clusters_xlf.cosmo['h']
fig, ax  = plt.subplots(figsize=(15, 10))

zarr = np.linspace(0.01, 1, 100)
_, dndz_all = clusters_xlf.dNdz(zarr = zarr, Slim = Slim_ext_soft, plot_ax =  ax, M_min=2.5e13)


_, dndz_massive = clusters_xlf.dNdz(zarr = zarr, Slim = Slim_ext_soft,M_min=5e13, plot_ax =  ax, lw = 3)


_, dndz_massive2 = clusters_xlf.dNdz(zarr = zarr, Slim = Slim_ext_soft,M_min=1e14, plot_ax =  ax, lw = 3)

ax.legend()
ax.set_yscale('linear')

n_obj = integrate.simps(dndz_massive, zarr)
print(f'M>5e13 (4.4e-14 Slim): n_obj in 0.658 of sky = {0.658*n_obj*(4*np.pi*(180/np.pi)**2)}')

n_obj = integrate.simps(dndz_massive2, zarr)
print(f'M>1e14 (4.4e-14 Slim): n_obj in 0.658 of sky = {0.658*n_obj*(4*np.pi*(180/np.pi)**2)}')

fig.savefig(path2plots + 'Clusters_dndz.png')
```

### LogNlogS

```python
#%matplotlib auto
clusters_xlf = ClustersXrayLuminosityFunction()

zarr = np.linspace(0.05, 1, 100)
fig,  ax =  plt.subplots( figsize = (10,7), sharex=False)

Sarr = np.geomspace(3e-14, 1e-10, 20)
for Mmin in [5e13, 7e13, 1e14]:
    print(f'Mmin = {Mmin}')
    _, logNlogS = clusters_xlf.logNlogS(Sarr, zarr = zarr, M_max=5e16, M_min=Mmin)
    ax.loglog(Sarr, logNlogS, label=f'Mmin = {Mmin:.0e} Msun/h', lw = 3)


ax.set_xlabel('F 0.5-2 keV, cgs')
ax.set_ylabel('N(>F), deg^-2')
ax.legend()
fig.tight_layout()
fig.savefig(path2plots+'Clusters_logNlogS.png')
```

## Linear bias factor

```python
zarr = np.linspace(0.05, 2, 100)
zarr, b_eff = clusters_xlf.b_eff(zarr, Slim = Slim_ext_soft, M_min = 5e13, M_max = 2e15)

h = clusters_xlf.cosmo['h']

fig, ax  = plt.subplots(figsize=(10, 5))
plt.plot(zarr, b_eff, label=f'b_eff, {Slim_ext_soft=}, M_min = 5e13 Msun/h', lw = 3)


plt.xlabel('z')
plt.ylabel('b(z)')
plt.legend()

fig.savefig(path2plots+'Clusters_b_eff.png')

```

## Mass distribution of clusters

```python
clusters_xlf = ClustersXrayLuminosityFunction(cosmo=ccl.Cosmology(**{'Omega_c': 0.25, 'Omega_b': 0.05,
                 'h': 0.7, 'sigma8':0.8, 'n_s': 0.96,
                 'transfer_function': 'boltzmann_camb',
                   'baryons_power_spectrum': 'nobaryons',
                   'matter_power_spectrum': 'linear'}))

zarr_clu = np.linspace(0.25,2, 1000)
Mmins = np.geomspace(1e14, 2e15, 15)
Slim = 4.4e-14

N_src_mass = []
for Mmin in Mmins: 
    print(f'Mmin = {Mmin}')
    n_src = clusters_xlf.logNlogS(Sarr = np.array([Slim]), zarr = zarr_clu, M_min=Mmin, M_max = 1e15*10)[1][0]

    N_src_mass.append(n_src)


N_src_mass = np.array(N_src_mass)


fig,  ax =  plt.subplots( figsize = (12,12))
plt.loglog((Mmins/1e14)/0.7, N_src_mass, label='My prediction, z >  0.25')

plt.xlabel('M500c, 1e14 Msun')
plt.ylabel('N(>M), deg^-2')
plt.legend()
plt.grid(which='both')


fig.savefig(path2plots+'Clusters_mass_distribution.png')

```
